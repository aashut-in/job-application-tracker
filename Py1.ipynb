```python
#OPERATORS

a = 100
b = 30

print(f"Variables: a = {a}, b = {b}\n")

print("--- Arithmetic Operators ---")
print(f"Sum (a + b): {a + b}")
print(f"Difference (a - b): {a - b}")
print(f"Product (a * b): {a * b}")
print(f"Quotient (a / b): {a / b}")
print(f"Floor Division (a // b): {a // b}")
print(f"Remainder (a % b): {a % b}")
print(f"Exponent (a ** 2): {a ** 2}\n")

print("--- Comparison Operators ---")
print(f"Equal (a == b): {a == b}")
print(f"Not Equal (a != b): {a != b}")
print(f"Greater Than (a > b): {a > b}")
print(f"Less Than (a < b): {a < b}")
print(f"Greater Than or Equal To (a >= b): {a >= b}")
print(f"Less Than or Equal To (a <= b): {a <= b}")
```

```python
# DATATYPES

# Integer (int): Represents whole numbers, positive or negative.
a = 12
print(f"Type of a: {type(a)}")

# String (str): Represents sequences of characters.
b = "yay"
print(f"Type of b: {type(b)}")

# Float (float): Represents real numbers with floating-point precision.
c = 10.00
print(f"Type of c: {type(c)}")

# Boolean (bool): Represents truth values (True or False).
d = True
print(f"Type of d: {type(d)}")

# List (list): An ordered, mutable collection of items.
my_list = [1, 2, 3, "hello"]
print(f"Type of my_list: {type(my_list)}")

# Tuple (tuple): An ordered, immutable collection of items.
my_tuple = (1, 2, "world")
print(f"Type of my_tuple: {type(my_tuple)}")

# Dictionary (dict): An unordered collection of key-value pairs.
my_dict = {"name": "Daisy"}
print(f"Type of my_dict: {type(my_dict)}")

# Set (set): An unordered collection of unique items.
my_set = {1, 2, 3, 3, 4}
print(f"Type of my_set: {type(my_set)}")
```

```python
# TYPECASTING

# --- Implicit Typecasting (Coercion) ---
# Python automatically converts one data type to another when it makes sense.
# For example, when an integer and a float are involved in an arithmetic operation.
int_num = 10
float_num = 5.5
result_add = int_num + float_num # int_num is implicitly converted to float
print(f"Implicit conversion (int + float): {int_num} + {float_num} = {result_add} (Type: {type(result_add)})\n")

# --- Explicit Typecasting (Type Conversion) ---
# Manually converting one data type to another using built-in functions.

# 1. To Integer (int())
str_to_int = "123"
print(f"String to Integer: int('{str_to_int}') = {int(str_to_int)} (Type: {type(int(str_to_int))})")
float_to_int = 10.75
print(f"Float to Integer: int({float_to_int}) = {int(float_to_int)} (Type: {type(int(float_to_int))})\n")

# 2. To Float (float())
int_to_float = 50
print(f"Integer to Float: float({int_to_float}) = {float(int_to_float)} (Type: {type(float(int_to_float))})")
str_to_float = "98.6"
print(f"String to Float: float('{str_to_float}') = {float(str_to_float)} (Type: {type(float(str_to_float))})\n")

# 3. To String (str())
int_to_str = 100
print(f"Integer to String: str({int_to_str}) = '{str(int_to_str)}' (Type: {type(str(int_to_str))})")
bool_to_str = True
print(f"Boolean to String: str({bool_to_str}) = '{str(bool_to_str)}' (Type: {type(str(bool_to_str))})\n")

# 4. To Boolean (bool())
# bool() evaluates empty strings, 0, None, empty lists/tuples/dicts/sets as False.
# Everything else is True.
print(f"Integer 0 to Boolean: bool(0) = {bool(0)}")
print(f"Integer 1 to Boolean: bool(1) = {bool(1)}")
print(f"Empty string to Boolean: bool('') = {bool('')}")
print(f"Non-empty string to Boolean: bool('hello') = {bool('hello')}")
print(f"Empty list to Boolean: bool([]) = {bool([])}")
print(f"Non-empty list to Boolean: bool([1]) = {bool([1])}\n")

# 5. To List (list())
str_to_list = "Python"
print(f"String to List: list('{str_to_list}') = {list(str_to_list)} (Type: {type(list(str_to_list))})")
tuple_to_list = (1, 2, 3)
print(f"Tuple to List: list({tuple_to_list}) = {list(tuple_to_list)} (Type: {type(list(tuple_to_list))})\n")

# 6. To Tuple (tuple())
list_to_tuple = [4, 5, 6]
print(f"List to Tuple: tuple({list_to_tuple}) = {tuple(list_to_tuple)} (Type: {type(tuple(list_to_tuple))})")
str_to_tuple = "abc"
print(f"String to Tuple: tuple('{str_to_tuple}') = {tuple(str_to_tuple)} (Type: {type(tuple(str_to_tuple))})\n")

# 7. To Set (set())
list_to_set = [1, 2, 2, 3, 4]
# Note: Sets automatically remove duplicate elements.
print(f"List to Set: set({list_to_set}) = {set(list_to_set)} (Type: {type(set(list_to_set))})")
str_to_set = "banana"
print(f"String to Set: set('{str_to_set}') = {set(str_to_set)} (Type: {type(set(str_to_set))})\n")
```
